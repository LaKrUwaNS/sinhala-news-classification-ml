```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

url = "https://hirunews.lk/news_listing.php?category=General"
driver.get(url)
time.sleep(3)
```

```python
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    new_height = driver.execute_script("return document.body.scrollHeight")

    # Stop when no more scrolling
    if new_height == last_height:
        break

    last_height = new_height

print("Scrolling completed")
```

```python
cards = driver.find_elements(By.CSS_SELECTOR, "#news-cards a.card-v4.listed")
print("Total articles found:", len(cards))
```

```python
data = []

for card in cards:
    title = card.find_element(By.CSS_SELECTOR, "h4.title").text
    description = card.find_element(By.CSS_SELECTOR, ".text-content-wrp .description").text
    date = card.find_element(By.CSS_SELECTOR, ".content-top > .description").text
    image_url = card.find_element(By.TAG_NAME, "img").get_attribute("src")
    link = card.get_attribute("href")

    data.append({
        "title": title,
        "description": description,
        "date": date,
        "image_url": image_url,
        "news_url": link
    })
```

```python
df = pd.DataFrame(data)
df.to_csv("hiru_news_raw.csv", index=False, encoding="utf-8-sig")

df.head()
```
